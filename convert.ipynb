```python
import subprocess
from pathlib import Path

# 🔧 Update if needed
VLC_PATH = r"D:\Program Files\VideoLAN\VLC\vlc.exe"

# 🎯 Target Sample Rate
TARGET_SR = 16000

def convert_mpeg_to_wav(input_path, output_path):
    input_file = Path(input_path)
    output_file = Path(output_path)

    if not input_file.exists():
        print("❌ Input file not found")
        return

    output_file.parent.mkdir(parents=True, exist_ok=True)

    cmd = [
        VLC_PATH,
        "--intf", "dummy",
        "--quiet",
        str(input_file),
        "--sout",
        f"#transcode{{acodec=s16l,channels=1,samplerate={TARGET_SR}}}:std{{access=file,mux=wav,dst={output_file}}}",
        "vlc://quit"
    ]

    try:
        subprocess.run(cmd, check=True)
        print(f"✅ Converted successfully → {output_file}")
    except subprocess.CalledProcessError as e:
        print(f"❌ Conversion failed: {e}")

# 🎧 Example usage
convert_mpeg_to_wav(
    r"D:\d_drive_project\Realtime_fraud_call_detection\wake_words\wake_words\SATYA_SCAM.aac",
    r"D:\d_drive_project\Realtime_fraud_call_detection\processed_dataset\SATYA_SCAM.wav"
)
```
